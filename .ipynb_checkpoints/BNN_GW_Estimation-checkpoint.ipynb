# Problem Statement



## Background 

Here we want to apply the similar idea from the linear regression model to the GW parameter estimation.

We make a simple comparison:

Linear regression model: $y = ax + b +\epsilon(x)$

GW parameter estimation model: $y = f(x) +\epsilon(x)$

Assume we use a model to approximate $f(x)$. And the model is parameterized by $\hat{f}_\tau(x)$ with parameters denoted as $\tau$

Enssentially we are trying to solve:

$$\max_{\theta, w}p(y|x; \theta, w) = \int_{\tau}{\color{red}{p(y| \tau, x; \theta)}} {\color{blue}{p_w(\tau | x)}} d\tau$$

The only change here is we do not have linear model assumption, instead, we use a neural network (nonlinear) model $\hat{f}_\tau(x)$ and replace the previous parameters $a,b$ with $\tau$.
 
## Dataset description

There are three datasets that include whitened waveforms.
The whitening is based on the power spectrol density calculated from real LIGO noise. The dataset includes 1-s long clean waveforms sampled at 8192Hz, generated via PyCBC package. Each of the waveforms includes 5 parameters (primary mass, secondary mass, final spin and quasi-normal modes (real and imaginary parts)). All of the data have maximum amplitude to be 1.

General information of the three datasets:

Training dataset:
1. 37512 waveforms of length 8192
2. 37512 X 5 labels

Testing dataset:
1. 6252 waveforms of length 8192
2. 6252 X 5 labels

Validation dataset:
1. 6252 waveforms of length 8192
2. 6252 X 5 labels

During the training, we need to simulated the noise environment. In this tutorial, we use Gaussian noise. However, this can be generalized to real LIGO noise as well. The noise level is defined as peak Signal-to-Noise ratio (pSNR), which is defined as:

$$\text{pSNR} = \frac{\text{maximum amplitude of waveform}}{\text{standard deviation of noise}}$$


## Model performance
Here we trained a model based on the training data and measure the uncertainty of the testing data given different pSNRs. 

P.S. Here to make illustration, the training is based on part of the training data and evaluation is based on another part of the training data. The two parts for training and testing do not have overlaps. This is just for illustration. The datasets, however, are fully prepared for future use. In this tutorial, we only consider estimating the final spin. The same approach can be applied to any physical parameters of interest.

# Reload model and make prediction

Because training a model might take time. Here we only show a trained model, without any hyperparameter tunnings. A model design script will also be presented for DIY use.

```python
###################################
# retrieve path for model weights
###################################

from pathlib import Path
export_dir = 'saved_model7'
subdirs = [x for x in Path(export_dir).iterdir()
           if x.is_dir() and 'temp' not in str(x)]
latest = str(sorted(subdirs)[-1])
```

```python
#################
# load model
#################

loaded = tf.saved_model.load(latest)
#print(list(loaded.signatures.keys()))  # ["serving_default"]
infer = loaded.signatures["serving_default"]
```

```python
##################
# Inference on final spin
##################
# This is for illustration purposes, one can use the model to infer other parameters as well.

check_SNR = 3.0
TEST_DATA_PATH = "final_whitened_data_with_labels_combined.h5"
BATCH_SIZE = 32
one_batch = tester_fixed_SNR(filename=TEST_DATA_PATH, shuffle=False,
                             batch_size = BATCH_SIZE, SNR=check_SNR)
tmp_data, tmp_label = one_batch
out = infer(tf.constant(tmp_data))[u'predictions'].numpy()
```

```python
check_idx = 7
_ = plt.hist(out[check_idx].squeeze(), bins=30)
plt.plot(tmp_label[check_idx][0].numpy(), 0., 'o', zorder=15,
         clip_on=False, markeredgecolor='k', color='#f97306')
plt.legend(['ground truth', 'estimations'])
plt.title('final spin estimation')
```

```python
####################################
# Inference across different SNRs
####################################

'''
check different SNRs, the output standard error of the estimation will change accordingly
'''

check_SNR = 1.
TEST_DATA_PATH = "final_whitened_data_with_labels_combined.h5"
BATCH_SIZE = 32
    
one_batch = tester_fixed_SNR(filename=TEST_DATA_PATH, shuffle=False,
                             batch_size = BATCH_SIZE, SNR=check_SNR)
tmp_data, tmp_label = one_batch
out = infer(tf.constant(tmp_data))[u'predictions'].numpy()

print('pSNR: {}, standard deviation: {}'.format(check_SNR, np.mean(np.std(out, 1), 0)))
```
